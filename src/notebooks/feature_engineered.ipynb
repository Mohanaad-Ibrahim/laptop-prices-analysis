# for save data دا اخر حاجا عملناها 


```python
df["Cpu"]
```

```python
def get_cpu_core(value: str):
    words = value.split(" ")
    for i, w in enumerate(words):
        if w.lower() == "core":
            return " ".join(words[i:i+2])
    return None
```

```python
df.insert(
    10,
    "cpu_core",
    df["Cpu"].apply(get_cpu_core)
)
```

```python
df
```

```python
def get_cpu_Memory(value: str):

    value = value.split(" ")[0]
    return value
get_cpu_Memory("256GB SSD")
```

```python
df.insert(
    13,
    "Memory_size",
    df["Memory"].apply(get_cpu_Memory)
)
```

```python
def get_cpu_Gpu(value: str):

    value = value.split(" ")[0] 
    return value
get_cpu_Gpu("Intel HD Graphics 520")
```

```python
def get_Gpu(value: str):
    keywords = ["Intel", "AMD", "NVIDIA", "HD", "UHD", "Radeon", "GeForce", "Iris"]
    words = value.split(" ")
    result = [w for w in words if w in keywords]
    return " ".join(result)
```

```python
df.insert(
    15,
    "type_Gpu",
    df["Gpu"].apply(get_Gpu)
)
```
